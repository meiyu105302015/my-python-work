```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
boston_dataset = load_boston()
```

```python
boston_dataset.feature_names
```

```python
# 載入數據
boston=pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names)
boston.head()
```

```python
# MEDV設為"自住房的平均房價，以千美元計"
boston["MEDV"]=boston_dataset.target
boston.head()
```

```python
boston.describe()
```

## 標準化

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD','TAX', 'PTRATIO', 'B', 'LSTAT']
boston[feature_names] = scaler.fit_transform(boston[feature_names])
boston.head()
```

## 變數相關性

```python
boston.corr()
cor=boston.corr()
plt.figure(figsize=(16,8))
sns.heatmap(cor,cmap='Blues',annot=True)
```

### 依相關性選取大於正負 0.4 的變數作為預測房價的指標
1. INDUS(城鎮非零售商用土地的比例)
2. NOX(一氧化氮濃度)
3. RM(住宅平均房間數)
4. TAX(每10000美元的全值財產稅率)
5. PTRATIO(城鎮師生比例)
6. LSTAT(人口中地位低下者的比例)

### 因此我選取上述六個因素來預測房價。

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
reg_boston = boston[['INDUS','NOX','RM','TAX','PTRATIO','LSTAT','MEDV']]
reg_boston
```

```python
X = reg_boston.loc[:, 'INDUS':'LSTAT'].values
Y = reg_boston.MEDV
```

```python
# 分割資料
x_train, x_test, y_train, y_test = train_test_split(X, Y, 
                                                    test_size=0.2,
                                                    random_state=0)
```

## 數據分析三部曲
### step 1. 開函數學習機

```python
model = LinearRegression()
```

### step 2. 訓練

```python
model.fit(x_train,y_train)
```

### step 3. 預測

```python
y_predict = model.predict(x_test)
```

```python
# 在線性上方的是高估
sns.set()
plt.scatter(y_test, y_predict)
plt.xlim(0,55)
plt.ylim(0,55)
plt.xlabel('Actual MEDV')
plt.ylabel('Predict MEDV')
plt.plot([0,55], [0,55], "r")
```
