```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 讀進上市公司營益分析資料

```python
df=pd.read_csv("http://mopsfin.twse.com.tw/opendata/t187ap17_L.csv")
```

### 查看資料型態

```python
df.info     #除了公司名稱，其他都是數值
```

```python
df
```

```python
df.sort_values(by="營業收入(百萬元)",ascending=False)
```

### 營業收入最高的五間公司

```python
df1_sort=df.sort_values(by="營業收入(百萬元)",ascending=False)
df1_sort.index=range(1,910)
df1_sort.head()
```

### 統一排名第10

```python
df1_sort[df1_sort.公司名稱=="統一"]
```

### 毛利率最高的五間公司

```python
df2_sort=df.sort_values(by="毛利率(%)(營業毛利)/(營業收入)",ascending=False)
df2_sort.index=range(1,910)
df2_sort.head()
```

### 統一的毛利率排名第167

```python
df2_sort[df2_sort.公司名稱=="統一"]
```

### 稅後純益率最高的五間公司

```python
df3_sort=df.sort_values(by="稅後純益率(%)(稅後純益)/(營業收入)",ascending=False)
df3_sort.index=range(1,910)
df3_sort.head()
```
