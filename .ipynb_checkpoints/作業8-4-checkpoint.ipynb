```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 分析大學僑生分佈狀況

```python
df=pd.read_csv("http://stats.moe.gov.tw/files/ebook/oversea/107/107oversea_B1-1a.csv")
```

```python
df
```

### 政大僑生人數

```python
df[df.學校名稱=="國立政治大學"]
```

```python
df["總人數"]=df.新生+df.舊生
```

```python
df
```

## 不同國家僑生人數最多的學校

```python
df1_sort=df.sort_values(by="總人數",ascending=False)
df1_sort.index=range(1,1261)
df1_sort.head()
```

## 日本人在台灣學校的人數

```python
df2_sort=df1_sort[df1_sort.僑居地=="日本"]
df2_sort.index=range(1,69)
df2_sort
```
