```python
# 互動函數
interact(f,x=3)
```

```python
# 可以隨便打字
interact(f,x="hi")
```

```python
def pipi2(x):
    print("拍拍")
```

```python
# 還沒輸入前就拍拍了
interact(pipi2,x="請輸入你的訊息")
```

```python
from ipywidgets import interact_manual
```

```python
# 輸入完文字才回覆
interact_manual(pipi2,x="請輸入你的訊息")
```

```python
def move(n=1):
    print(" "*n+"(*´▽`*)")
# n=1 互動的預設值在1，如果沒設，預設值在中間
```

```python
move(20)
```

```python
interact(move,n=(1,40))
```

```python
# 少掉最後一行
interact(move,n=(1,40));
```

```python
def myplot(n=1):
    x=np.linspace(-5,5,100)
    y=np.sinc(n*x)
    plt.plot(x,y)
```

```python
interact(myplot,n=(1,20));
```

## while 迴圈
條件成立就執行，不成立就不執行

```python
s=0
while s<10:
    print(s)
    s=s+1
```

```python
msg="多啦A夢"
while(msg!="bye"):
    msg=input(">>")
    print("拍拍")
```
