```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

## 用 contour繪製等高線

```python
def f(x,y):
    return np.exp(-x**2)*np.exp(y)
```

```python
a, b = plt.subplots()
b.set_xlabel('x', fontsize=16)
b.set_ylabel('y', fontsize=16)
n = 200
x = np.linspace(-2,2,n)
y = np.linspace(-2,2,n)
X,Y = np.meshgrid(x, y)
line_lv = [0.05,0.3,1,2,4,6]
cf_lv = [0,0.05,0.3,1,2,4,6,np.exp(2)]
lb_pos = [(0,1.8),(0,1.4),(0,0.8),(-0.75,0.7),(0.75,0.7),(-1,0),(1,0),(-1.5,-1),(1.5,-1)]
plt.contourf(X, Y, f(X, Y), levels=cf_lv, alpha=.75, cmap='coolwarm')
C = plt.contour(X, Y, f(X, Y), levels=line_lv, colors='black', linewidth=.5)
plt.clabel(C, inline=True, fontsize=10, manual=lb_pos)
```
