## 分析大專校院各校科系別概況

```python
df=pd.read_csv("http://stats.moe.gov.tw/files/detail/107/107_sdata.csv")
```

```python
df
```

## 學生數最多的學校及科系

```python
df1_sort=df.sort_values(by="學生數",ascending=False)
df1_sort.index=range(1,9907)
df1_sort.head()
```

### 一般大學中學生數最多的學系及科系

```python
df2_sort=df1_sort[df1_sort.體系別=="1 一般"]
df2_sort.index=range(1,5401)
df2_sort.head()
```

### 政大學生最多的系

```python
df3_sort=df1_sort[df1_sort.學校名稱=="國立政治大學"]
df3_sort.index=range(1,158)
df3_sort.head()
```

### 政大老師最多的系

```python
df4_sort=df.sort_values(by="教師數",ascending=False)
df4_sort=df4_sort[df4_sort.學校名稱=="國立政治大學"]
df4_sort.index=range(1,158)
df4_sort.head()
```
