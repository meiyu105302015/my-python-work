```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 讀進上市公司基本資料

```python
df=pd.read_csv("http://mopsfin.twse.com.tw/opendata/t187ap03_L.csv")
```

```python
df
```

## 取出公司代號、公司簡稱等想要的資訊

```python
df1=df[["公司代號","公司名稱","公司簡稱","產業別","董事長","總經理","發言人","成立日期","上市日期","普通股每股面額","實收資本額","私募股數","特別股","網址"]]
```

```python
df1
```

## 實收資本額最大的五間公司

```python
df2_sort=df1.sort_values(by="實收資本額",ascending=False)
df2_sort.index=range(1,952)
df2_sort.head()
```

### 金融保險業中實收資本額最高的五間公司

```python
df3_sort=df2_sort[df2_sort.產業別=="金融保險業"]
df3_sort.index=range(1,36)
df3_sort.head()
```
