## 微軟股票分析

```python
import pandas_datareader.data as web
```

```python
df=web.DataReader("MSFT","yahoo",start="2015-1-1",end="2019-12-12")
```

### 前五筆微軟股票資訊

```python
df.head()
```

```python
#收盤價折線圖
df.Close.plot()
```

### 計算報酬率

```python
P=df['Adj Close']
```

```python
r=P.diff()/P
```

```python
r[-100:].plot()
```

## 移動平均
1. 每日收盤價
2. 20天的移動平均
3. 60天的移動平均

```python
P.plot()
P.rolling(window=20).mean().plot()
P.rolling(window=60).mean().plot()
```
